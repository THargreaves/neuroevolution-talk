# Evolutionary Algorithms

## Evolution Recap

- An organism/individual is differentiated by its unique genetic code
- This genetic code will express itself in physical attributes
- These attributes impact the likelihood of survival
- Surviving organisms are more likely to reproduce
- Reproduction allows an organism to partially pass on their genes
- Mutations can occur at any time to change an organism's genes

## Components of Evolution

- Genes
- Fitness/Selection
- Cross-over/Reproduction
- Mutation

### Genes

- The smallest unit of selection
- Possible expressions of a particular gene are called alleles
- Contributes to the physical attributes of the organism

### Fitness/Selection

- Each organism has an abstract value of fitness
- This relates to its chance of survival/reproduction
- Genes that encourage fitness are more likely to disseminate

### Cross-over/Reproduction

- Two organisms can reproduce to have children
- Each child will take half their genetic material from each parent
- This loosely corresponds to 'mixing' the genes of the parents

## Mutation

- Genes can mutate to change their expression
- This can occur at any time but is more likely at birth

## Example: Quadratic Regression

````python
# Parameters
A = -1  # quadratic coeff
B = 1  # linear coeff
C = 2  # intercept
N = 20  # number of samples
S2 = 0.1  # variance of errors

# Create data
np.random.seed(1729)
x = np.random.normal(1, 1, N)
y = A * x**2 + B * x + C + np.random.normal(0, np.sqrt(S2), N)
````

### Task

- Find optimal (least squares) coefficents to fit noisy quadratic data

````python
fig, ax = plt.subplots()
ax.scatter(x, y, c='k')
x_new = np.linspace(np.min(x), np.max(x), 100)
y_new = A * x_new**2 + B * x_new + C
ax.plot(x_new, y_new, c = 'red',
        label = 'Ground Truth ($-x^2+x+2$)')
fig.legend()
plt.show()
````

### Cross-over/Reproduction

```python
for gene in mother.population.gene_pool.keys():
    # Convex combination of parents' alleles
    p = random.random()
    genome[gene] = p * mother.genome[gene] + \
        (1-p) * father.genome[gene]
```

### Mutation

```python
for gene, alleles in self.population.gene_pool.items():
    if self.population.mutation_chance > random.random():
        # Convex combination of current allele and random
        # choice from the gene pool
        p = (1 + random.random()) / 2
        self.genome[gene] = p * self.genome[gene] + \
            (1-p) * random.uniform(alleles[0], alleles[1])
```

## Results

````python
sys.path.insert(0, '../quadratic_regression')
import quadratic_regression

# Parameters
P = 20  # population size
G = 20  # number of generations
GP = {
    'a': (-3, 3),
    'b': (-3, 3),
    'c': (-3, 3),
}  # gene pool
MC = 0.05  # mutation chance
RP = 0.4  # retained proportion
RS = 0.1  # chance of random selection
GT = (x, y)  # ground truth
SEED = 42
````

````python
# Setup plotting area
fig, ax = plt.subplots()
ax.scatter(x, y, c='black')

lines = []
for __ in range(P):
    lines.append(ax.plot([], [], lw=2, alpha=0.5)[0])

# Create population
population = quadratic_regression.Population(P, GP, MC, RP, RS, GT, SEED)
    
# Animation functions
def init():
    for i in range(P):
        lines[i].set_data([], [])
    return lines
    
def animate(f):
    for i, org in enumerate(population.organisms):
        y_new = org.genome['a'] * x_new ** 2 + \
            org.genome['b'] * x_new + \
            org.genome['c']
        lines[i].set_data(x_new, y_new)
    population.evolve()
    return lines
    
# Animate
anim = ani.FuncAnimation(fig, animate, init_func=init,
                         frames=G, interval=1000, blit=True)
plt.close()
anim
````

````python
# Setup plotting area
fig, axs = plt.subplots(3, 1, sharex=True)
for i, (lab, val) in enumerate(zip(('$a$ ($x^2$)', '$b$ ($x$)', 'c ($1$)'),
                                   (A, B, C))):
    axs[i].axhline(val, c='black')
    axs[i].set_ylabel(lab)
    axs[i].set_ylim(val-2, val+2)
axs[2].set_xticks(np.linspace(0, 20, 11))
axs[2].set_xlim(0, 20)
axs[2].set_xlabel("Generation")
fig.text(0, 0.5, 'Coefficient', va='center', rotation='vertical')
fig.tight_layout()

errors = []
lines = []
for i in range(3):
    errors.append(axs[i].fill_between([], [], []))
    lines.append(axs[i].plot([], [], lw=2)[0])

# Create population
population = quadratic_regression.Population(P, GP, MC, RP, RS, GT, SEED)

# Animation functions
lower = [[] for __ in range(3)]
mean = [[] for __ in range(3)]
upper = [[] for __ in range(3)]

def init():
    for i in range(3):
        axs[i].collections.clear()
        errors[i] = axs[i].fill_between([], [], [])
        lines[i].set_data([], [])
    return errors + lines
    
def animate(f):
    for i, g in enumerate(('a', 'b', 'c')):
        coeffs = np.empty((P,))
        for j, org in enumerate(population.organisms):
            coeffs[j] = org.genome[g]
        mu = np.mean(coeffs)
        sd = np.std(coeffs)

        mean[i].append(mu)
        # 95% normal confidence intervals
        lower[i].append(mu - 1.96 * sd / math.sqrt(P))
        upper[i].append(mu + 1.96 * sd / math.sqrt(P))
        
        lines[i].set_data(np.arange(f+1), mean[i])
        axs[i].collections.clear()
        errors[i] = axs[i].fill_between(np.arange(f+1), lower[i], upper[i],
                                        color='C0', alpha=0.2)
        
    population.evolve()
    return errors + lines

# Animate
anim = ani.FuncAnimation(fig, animate, init_func=init,
                         frames=G, interval=1000, blit=True)
plt.close()
anim
````

## Comparison to Grid Search

- Grid search on integer lattice: $7^3=343$ combinations
- For precision of $0.5$ this increases to $13^3=2197$
- Evolutionary method above had $20$ organisms per generation
- Linear in number of generations 
- Reasonable convergence at $5$ generations

## Comparison to Gradient Descent

````python
fig, ax = plt.subplots()
RES = 25
mse = np.empty((RES, RES))
for i, a in enumerate(np.linspace(-3, 3, RES)):
    for j, b in enumerate(np.linspace(-3, 3, RES)):
        y_hat = a * x**2 + b * x + C
        mse[i, j] = np.mean(np.square(y - y_hat))
pos = ax.imshow(mse, origin='lower')
ax.set_xlabel('$a$ ($x^2$)')
ax.set_xticks(np.linspace(0, RES-1, 7))
ax.set_xticklabels(np.linspace(-3, 3, 7))
ax.set_yticks(np.linspace(0, RES-1, 7))
ax.set_yticklabels(np.linspace(-3, 3, 7))
ax.set_ylabel('$b$ (x)')
cbar = fig.colorbar(pos, extend='both')
cbar.set_label('Mean Squared Error')
plt.plot((B + 3) * RES / 6, (A + 3) * RES / 6, 'w+')
plt.text((B + 3) * RES / 6, (A + 3) * RES / 6 + 0.5, 'Optimum',
         {'fontsize': 8}, c='w',
         horizontalalignment='center',
         verticalalignment='bottom')
plt.show()
````

````python
sys.path.insert(0, '../image_classification')
import image_classification

# Parameters
P = 20  # population size
G = 10  # number of generations
GP = {
    'num_neurons': [32, 64, 128, 256, 512, 768, 1024],
    'num_layers': [1, 2, 3, 4, 5],
    'use_bias': [True, False],
    'dropout_rate': [0., .1, .2, .3, .4],
    'activation': ['relu', 'elu', 'tanh', 'sigmoid'],
    'optimiser': ['rmsprop', 'adam', 'sgd', 'adagrad',
                  'adadelta', 'adamax', 'nadam'],
}  # gene pool
MC = 0.05  # mutation chance
RP = 0.4  # retained proportion
RS = 0.1  # chance of random selection
SEED = 253
````

````python
# Setup plotting area
fig, ax = plt.subplots()
ax.set_ylabel("Average Fitness (Accuracy)")
ax.set_ylim(0.5, 1)
ax.set_xticks(np.linspace(0, 10, 11))
ax.set_xlim(0, 10)
ax.set_xlabel("Generation")

line = ax.plot([], [], lw=2)[0]

# Create population
population = image_classification.Population(P, GP, MC, RP, RS, SEED,
                                             verbose=False)

# Animation functions
fitness = []

def init():
    line.set_data([], [])
    return [line]
    
def animate(f):
    fitness.append(population.average_fitness)
    line.set_data(np.arange(f+1), fitness)
        
    population.evolve()
    return [line]

# Animate
anim = ani.FuncAnimation(fig, animate, init_func=init,
                         frames=G, interval=1000, blit=True)
plt.close()
anim
````
